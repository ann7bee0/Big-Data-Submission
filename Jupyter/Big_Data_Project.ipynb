```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("commodity_trade_statistics_data.csv")
```

```python
df
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
for i in df:
    print("++++++++++++++++++++++++++++++++++++++++++++++++Column ",i,"+++++++++++++++++++++++++++++++++++++++++++++++++++")
    print(df[i].unique())
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    print()
```

```python
for i in df:
    print("++++++++++++++++++++++++++++++++++++++++++++++++Column ",i,"+++++++++++++++++++++++++++++++++++++++++++++++++++")
    print(df[i].value_counts())
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    print()
```

```python
df.rename(columns = {'country_or_area':'country'}, inplace = True)
df.drop(df[df['category'] == 'all_commodities'].index, inplace = True)
df.drop(df[df['category'] == '99_commodities_not_specified_according_to_kind'].index, inplace = True)
df.drop(df[df['country'] == 'EU-28'].index, inplace = True)
```

```python
df['trade_usd']=(df['trade_usd']/1000000).round(2)
df.head()
```

```python
df.info()
```

<hr>

```python
df.isna().sum()
```

```python
df['weight_kg'].fillna(df['weight_kg'].mean(), inplace=True)
df['quantity'].fillna(df['quantity'].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
for i in df.select_dtypes(include=['int64','float64']):
    print("++++++++++++++++++++++++++++++++++++++++++++++++Column ",i,"+++++++++++++++++++++++++++++++++++++++++++++++++++")
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, y=i, color='skyblue')
    plt.title(f'Box and Whisker Plot for {i}', fontsize=14)
    plt.ylabel(i, fontsize=12)
    plt.xlabel('')  # No x-label as it's not needed for single-variable plots
    plt.show()
    print("++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
    print()
```

```python
non_numeric_columns = df.select_dtypes(exclude=['int64', 'float64']).columns
```

```python
df_cleaned = df.dropna(subset=non_numeric_columns)
```

```python
df_cleaned
```

```python
df_cleaned.isna().sum()
```

```python
df_cleaned.drop(columns=['comm_code'], inplace=True)
```

```python
df_cleaned['weight_kg'] = df['weight_kg'] / 1000
```

```python
df_cleaned
```

```python
# Save the cleaned DataFrame to a CSV file
df_cleaned.to_csv('cleaned_data.csv', index=False)
```

```python
# df_cleaned.to_csv()
df = df_cleaned
```

```python
# Total trade value by country
total_trade_by_country = df_cleaned.groupby('country')['trade_usd'].sum().sort_values(ascending=False)
print(total_trade_by_country)
```

```python
# Top 5 countries by trade value
top_countries = total_trade_by_country.head(5)
print(top_countries)
```

```python
# Distribution of import vs export
trade_flow_distribution = df_cleaned['flow'].value_counts(normalize=True) * 100
print(trade_flow_distribution)
```

```python
# Average weight in tonnes by country
avg_weight_by_country = df_cleaned.groupby('country')['weight_kg'].mean().sort_values(ascending=False)
print(avg_weight_by_country)
```

```python
# Total weight contribution by commodity
weight_by_commodity = df_cleaned.groupby('commodity')['weight_kg'].sum().sort_values(ascending=False)
print(weight_by_commodity)
```

```python
# Average quantity per commodity
avg_quantity_per_commodity = df_cleaned.groupby('commodity')['quantity'].mean().sort_values(ascending=False)
print(avg_quantity_per_commodity)
```

```python
# Commodities with the highest total quantity
high_demand_commodities = df_cleaned.groupby('commodity')['quantity'].sum().sort_values(ascending=False)
print(high_demand_commodities)
```

```python
# Correlation between quantity and weight
correlation = df_cleaned[['quantity', 'weight_kg']].corr()
print(correlation)
```

```python
# Yearly trade value and weight trends
yearly_trends = df.groupby('year')[['trade_usd', 'weight_tonnes']].sum()
print(yearly_trends)
```

```python
# Trade value of each commodity over the years
commodity_trends = df.pivot_table(values='trade_usd', index='year', columns='commodity', aggfunc='sum')
print(commodity_trends)
```

```python
# Most traded category each year
top_categories_per_year = df.groupby(['year', 'category'])['trade_usd'].sum().reset_index()
top_categories_per_year = top_categories_per_year.loc[top_categories_per_year.groupby('year')['trade_usd'].idxmax()]
print(top_categories_per_year)
```

```python
# Top commodities by trade value
top_commodities = df.groupby('commodity')['trade_usd'].sum().sort_values(ascending=False)
print(top_commodities)
```

```python
# Total trade value by category
trade_by_category = df.groupby('category')['trade_usd'].sum()
print(trade_by_category)
```

```python
# Number of unique commodities per country
unique_commodities_by_country = df.groupby('country')['commodity'].nunique()
print(unique_commodities_by_country)
```

```python
# Import vs export balance for each country
trade_balance = df.pivot_table(values='trade_usd', index='country', columns='flow', aggfunc='sum', fill_value=0)
trade_balance['balance'] = trade_balance.get('export', 0) - trade_balance.get('import', 0)
print(trade_balance)
```

```python
# Top trading countries by trade value
top_trading_partners = df.groupby('country')['trade_usd'].sum().sort_values(ascending=False).head(5)
print(top_trading_partners)
```

```python
# Check missing data patterns
missing_data = df.isnull().sum()
print(missing_data)
```

```python
# Boxplot for outliers in trade_usd
plt.figure(figsize=(10, 5))
sns.boxplot(x=df['trade_usd'])
plt.title('Outliers in Trade Value')
plt.show()
```

```python
# Histogram for trade value
plt.figure(figsize=(10, 5))
sns.histplot(df['trade_usd'], kde=True, bins=30)
plt.title('Distribution of Trade Value')
plt.xlabel('Trade Value (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix of numeric columns
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
